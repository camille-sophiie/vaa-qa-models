## Dataset Creation

```python
#initialise lists for future df
tweets =[]
likes= []
time=[]
politician_id=[]
```

```python
for person in twitter_id:
    for i in tweepy.Cursor(api.user_timeline, id=person, tweet_mode="extended").items():
        tweets.append(i.full_text)
        likes.append(i.favorite_count)
        time.append(i.created_at)
        politician_id.append(i.user.screen_name)
```

```python
df = pd.DataFrame({"tweets": tweets, "likes": likes, "time":time, "politician_id":politician_id})
```
