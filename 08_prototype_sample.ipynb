```python
import pandas as pd
from transformers import pipeline
```

```python
df_questions = pd.read_csv("./csv/questions_model.csv")
```

```python
df_questions = df_questions[:5]
```

```python
df_questions
```

```python
df = pd.read_csv("./csv/df_translated_tweets.csv")
```

```python
#join all tweets

context_tweets = []
for i in range(len(df)):
    if isinstance(df.iloc[i]['translated_tweets'], float):
        continue
    else:
        context_tweets.append(df.iloc[i]['translated_tweets'][23:-3])

#join all the context tweets
context_tweets_all = ' '.join(context_tweets)    
```

```python
qa_model = pipeline("question-answering", model = "srcocotero/bert-large-qa")
```

```python
def qa(question_smartvote):
    return qa_model(question = question_smartvote , context = context_tweets_all)
```

```python
df_questions["huggingface_answers_simplified"] = df_questions["Simplified question"].apply(qa)
```
